# Scale × ODYSSEY — Exploratory Data Analysis
Visualise the dataset distribution and sample images before training.

```python
import os
import sys
sys.path.insert(0, '../src')

from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

DATA_ROOT = Path('../data')
CLASSES   = ['spiral', 'elliptical', 'nebula', 'star_cluster', 'planetary']
SPLIT     = 'train'
```

```python
# ── Class distribution ───────────────────────────────────────────────────────
counts = {}
for cls in CLASSES:
    cls_dir = DATA_ROOT / SPLIT / cls
    if cls_dir.exists():
        n = len(list(cls_dir.glob('*')))
    else:
        n = 0
    counts[cls] = n
    print(f'{cls:<18}: {n} images')

plt.figure(figsize=(8, 4))
plt.bar(counts.keys(), counts.values(), color='steelblue')
plt.title('Class Distribution (Train Split)')
plt.ylabel('Number of Images')
plt.xticks(rotation=15)
plt.tight_layout()
plt.savefig('../results/class_distribution.png', dpi=150)
plt.show()
```

```python
# ── Sample images from each class ─────────────────────────────────────────
fig, axes = plt.subplots(1, 5, figsize=(16, 4))

for ax, cls in zip(axes, CLASSES):
    cls_dir = DATA_ROOT / SPLIT / cls
    imgs = list(cls_dir.glob('*.jpg')) + list(cls_dir.glob('*.png'))
    if imgs:
        img = Image.open(imgs[0]).convert('RGB').resize((224, 224))
        ax.imshow(img)
    ax.set_title(cls.replace('_', '\n'), fontsize=10)
    ax.axis('off')

plt.suptitle('Sample Images per Class', fontsize=14)
plt.tight_layout()
plt.savefig('../results/sample_images.png', dpi=150)
plt.show()
```

```python
# ── Image resolution distribution ────────────────────────────────────────
widths, heights = [], []
for cls in CLASSES:
    for img_path in list((DATA_ROOT / SPLIT / cls).glob('*'))[:100]:
        try:
            img = Image.open(img_path)
            widths.append(img.width)
            heights.append(img.height)
        except:
            pass

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].hist(widths,  bins=30, color='coral');  axes[0].set_title('Image Widths')
axes[1].hist(heights, bins=30, color='teal');   axes[1].set_title('Image Heights')
plt.tight_layout()
plt.show()
```
